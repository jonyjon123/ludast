```python
import pandas as p
import numpy as np
```

```python
documents_categories = p.read_csv("documents_categories.csv")
```

```python
#too many values - cause memory error.
#for each document take category with max confidence value. if confidence_value identical for two categories or more - take all
max_rows = documents_categories.groupby(['document_id'])['confidence_level'].transform(max) == documents_categories['confidence_level']
categories = documents_categories[max_rows]
```

```python
#create category dummies
cat_dum = p.get_dummies(categories["category_id"], prefix="cat")
```

```python
#replace category_id column with it's dummies
categories.drop("category_id", axis=1, inplace=True)
categories = categories.join(cat_dum)
```

```python
#group rows with the same document and confidence level
categories = categories.groupby(['document_id', 'confidence_level']).sum().reset_index()
```

```python
del documents_categories, max_rows, cat_dum
```

```python
clicks_train = p.read_csv('clicks_train.csv')
```

```python
#use only part of clicks_train - otherwise cause memory error
train = clicks_train[:1000000]
```

```python
events = p.read_csv('events.csv',usecols=['display_id','document_id', 'platform'])
```

```python
#add events columns to train
train = train.merge(events, how='left', on='display_id')
```

```python
#add category dummies to train
train = p.merge(train, categories)
```

```python
promoted = p.read_csv("promoted_content.csv")
```

```python
#add category dummies to promoted content
promoted = promoted.merge(categories, how='left', on='document_id')
```

```python
#merge promoted content and train on ad_id
#now we have connection between ad category and document category
#can choose confidence level and slice
ad_doc_cat = p.merge(train, promoted, on=['ad_id'], suffixes=('_ad', '_doc'))
```
