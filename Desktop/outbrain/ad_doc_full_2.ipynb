```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('C:\\Users\\rludmer\\Downloads\\DataScienceWorkshop\\clicks_train.csv')
test = pd.read_csv('C:\\Users\\rludmer\\Downloads\\DataScienceWorkshop\\clicks_test.csv')
events = pd.read_csv('C:\\Users\\rludmer\\Downloads\\DataScienceWorkshop\\events.csv',usecols=['display_id','document_id'])
```

```python
train = train.merge(events, how='left')
test = test.merge(events, how='left')
del events
```

```python
ad_doc = train.groupby(['ad_id', 'document_id'])['clicked'].agg(['count','sum','mean']).reset_index()
```

```python
ad_doc_prob = train.groupby(['display_id']).size().reset_index(name="probability")
```

```python
ad_doc_prob["probability"] = (1 / ad_doc_prob["probability"]).astype(float)
```

```python
ad_doc_prob = train.merge(ad_doc_prob.filter(['display_id', 'probability']), how='left')
```

```python
ad_doc_prob2 = ad_doc_prob.groupby(["ad_id", "document_id", "probability"]).size().reset_index(name="count")
```

```python
ad_doc_prob3 = ad_doc_prob2[ad_doc_prob2["count"] > 2]
```

```python
ad_doc_prob = ad_doc_prob.groupby(["ad_id", "document_id"])["probability"].mean().reset_index(name="avg probability")
```

```python
ad_doc = pd.merge(ad_doc, ad_doc_prob, on =['ad_id','document_id'])
```

```python
del ad_doc_prob
```

```python
ad_doc['reg_percentage'] = (ad_doc['sum'] + 12*(ad_doc['mean'].mean())) / (12 + ad_doc['count'])
```

```python
ad_doc["ad_doc_strength"] = ad_doc["reg_percentage"] - ad_doc["avg probability"]
```

```python
ad_doc_strength = ad_doc.drop(ad_doc.columns[[2, 3, 4, 5, 6]], axis=1)
```

```python
ad_strength = train.groupby(['ad_id'])['clicked'].agg(['count','sum','mean']).reset_index()
```

```python
ad_strength["ad_strength"] = (ad_strength['sum'] + 12*(ad_strength['mean'].mean())) / (12 + ad_strength['count'])
```

```python
ad_strength = ad_strength.drop(ad_strength.columns[[1, 2, 3, 4]], axis=1)
```

```python
sub = test.merge(ad_strength, how='left')
```

```python
sub.ad_strength.fillna(ad_strength["ad_strength"].mean(), inplace=True)
```

```python
sub = sub.merge(ad_doc_strength, how='left')
```

```python
sub.ad_doc_strength.fillna(sub["ad_strength"], inplace=True)
```

```python
sub["strength"] = (0.5 * sub["ad_strength"]) + (0.5 * sub["ad_doc_strength"])
```

```python
final = sub.sort_values(['display_id','strength'], ascending=False)
```

```python
final.drop(final.columns[[2, 3, 4, 5]], axis=1, inplace=True)
```

```python
final = final.groupby('display_id').ad_id.apply(lambda x: " ".join(map(str,x))).reset_index()
```

```python
final.to_csv("ad_doc_full2.csv", index=False)
```
