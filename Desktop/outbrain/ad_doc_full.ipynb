```python
train = pd.read_csv("C:\\Users\\rludmer\\Downloads\\DataScienceWorkshop\\clicks_train.csv")
event = pd.read_csv("C:\\Users\\rludmer\\Downloads\\DataScienceWorkshop\\events.csv", dtype={"platform" : str})
```

```python
doc_id = train.merge(event.filter(['display_id', 'document_id']), how='left', left_on='display_id', right_on='display_id',
                           suffixes=('', '_'))
```

```python
doc_id_count = doc_id.groupby(['ad_id', 'document_id']).size().reset_index(name="count").sort_values(by="count")
```

```python
clicks = doc_id.groupby(['ad_id', 'document_id'])["clicked"].sum().reset_index(name="clicked").sort_values(by="clicked")
```

```python
doc_id_count = pd.merge(doc_id_count, clicks, on =['ad_id','document_id'])
```

```python
doc_id_count["percentage"] = (doc_id_count["clicked"] / doc_id_count["count"]).astype(float)
```

```python
doc_id_count.sort_values(by="percentage", ascending=False)
```

```python
doc_id_prob = train.groupby(["display_id"]).size().reset_index(name="probability")
```

```python
doc_id_prob["probability"] = (1 / doc_id_prob["probability"]).astype(float)
```

```python
doc_id_prob = doc_id.merge(doc_id_prob.filter(['display_id', 'probability']), how='left', left_on='display_id', right_on='display_id',
                           suffixes=('', '_'))
```

```python
doc_id_prob = doc_id_prob.groupby(["ad_id", "document_id"])["probability"].mean().reset_index(name="avg probability")
```
